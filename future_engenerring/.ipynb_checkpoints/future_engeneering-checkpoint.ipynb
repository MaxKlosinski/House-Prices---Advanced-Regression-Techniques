## Data imputation
It is not possible to perform imputation using a tree model because I would have to write it in an optiuna training loop that would run many times and significantly extend the model training time. Therefore, I will perform imputation using a simple method, i.e., filling in with the average, but divided into street groups to simulate the results, but only within districts.

## Why do I think that dividing data into district groups is better than dividing it into simple groups without any dependencies?
Because when I divide it into districts, it means that data from other districts will not affect other districts, and even though it would be possible to apply additional division conditions that would take into account dependencies not only between districts, I still consider this method to be correct in terms of the small amount of data, and additionally, other categories do not present a large number of categories, so I consider such a division to be unnecessary.

```python
# Important note when working with the pandas library. Tables in pandas have a special feature: if you only want to perform calculations on a specific column, you can 
# use [] immediately after the variable name, and this will only make changes to that column without unnecessary data separation.

train_full['LotFrontage'] = train_full['LotFrontage'].replace("none", np.nan)

mean_in_group = train_full.groupby('Neighborhood')['LotFrontage'].transform('mean')

train_full['LotFrontage'] = train_full['LotFrontage'].fillna(mean_in_group)

test['LotFrontage'] = test['LotFrontage'].replace("none", np.nan)

mean_in_group = test.groupby('Neighborhood')['LotFrontage'].transform('mean')

test['LotFrontage'] = test['LotFrontage'].fillna(mean_in_group)
```

## Removing outliers
At first, I wanted to remove them using MDS, but it turns out that it has too much impact on the data I need.
That's why I'll also use quartile removal, because it's not based on the median but on quartiles, which will give a less strict threshold for cutting skewed values, but this will become clear once I figure out how to mask values, because it turns out that pandas is clever but did not anticipate that someone would be very lazy and would not want to write a whole long function for grouping and then joining, but you can filter and, based on a True/False mask, you can unmask the deleted rows.

I remove outliers after dividing into districts and calculate the maximum and minimum values in a given group, and I remove values above these.
I group by district because the SalePrice data has the highest correlation with this data.

```python
def find_values_in_the_Quartile_interval(x):
    
    if x._name == 'StoneBr' or x._name == 'OldTown' or x._name == 'NoRidge' or x._name == 'NridgHt':
        
        Q1 = x.quantile(0.25)
        Q3 = x.quantile(0.75)
        
        IQR = Q3 - Q1
        
        min_value = Q1 - 1.5 * IQR
        max_value = Q3 + 1.5 * IQR
    
        # Important note: the word “and” does not work the same way as the & symbol.
        is_in_range = (x >= min_value) & (x <= max_value)
    else:
        is_in_range = x == x
    
    return is_in_range
```

```python
# The given function will return a mask of binary values indicating where our condition was met, and based on that, I will filter my group.
# This is possible thanks to the groupby() function, which allows us to return our entire table in the correct order.
# To be used in our general, ungrouped table.
mask = train_full.groupby('Neighborhood')['SalePrice'].transform(find_values_in_the_Quartile_interval)

# To apply the mask, simply enter your mask in square brackets and you're done.
train_full = train_full[mask]
print(train_full['SalePrice'])
```

## Basic statistical operations
The maximum and minimum values, mean, median, and standard deviation will be calculated for the data set from the sale price column in order to determine how widely the data is spread, and the median will be used to see how large the deviation from the center is. The standard deviation will indicate the actual deviations of the data from the mean. The standard deviation will be expressed as a **coefficient of variation** because the standard deviation will represent the data but in the same unit as the data we operated on, and the coefficient of variation will allow us to convert the given quantity into the quantity of data but expressed as a percentage. In other words, it will tell us the percentage deviation from the mean.

```python
print(train_full["SalePrice"].median())
print(train_full["SalePrice"].mean())
mean_SalePrice = train_full["SalePrice"].mean()
std_SalePrice = train_full["SalePrice"].std()
max_SalePrice = train_full["SalePrice"].max()
min_SalePrice = train_full["SalePrice"].min()

print()
print("Standard deviation expressed as a coefficient of variation.")
print((std_SalePrice/mean_SalePrice) * 100)
print("Normal STD:")
print(std_SalePrice)

# Uwaga!!! Regóła empiryczna jest w tym wypadku tylko w gwoli ciekawostki bo tak naprawdę nie mogę jej zastosować bo dane są bocno skośne.
print()
print(f"max - min: {max_SalePrice - min_SalePrice}")
print("Check empirical rule 99.7:")
print(f"{mean_SalePrice + 3 * std_SalePrice}")
```

This is necessary because there are as many as 81 missing data points in this column, and as the correlogram shows, this column is important for the prediction given by my model.

## Searching for normally distributed data using the Shapiro-Wilk test

```python
Zbiór_wartości_o_rozkładzie_normalnym = []
Odrzucone_zbiory = []

print("zbiory dla shapiro-wilk")
for kolumn in train_to_analise.drop(columns=categorical_columns):
    shapiro_test = stats.shapiro(train_to_analise[kolumn], nan_policy='omit')
    if kolumn == 'count_of_porch':
        cos = 0
    if shapiro_test.pvalue > 0.05:
        Zbiór_wartości_o_rozkładzie_normalnym.append(kolumn)
    else:
        Odrzucone_zbiory.append(kolumn)

print()
print("Zbiory o rozkładzie normalnym")
print(Zbiór_wartości_o_rozkładzie_normalnym)
print()
print("Zbiory odrzucone o nie o rozkładzie normalnym")
print(Odrzucone_zbiory)

print("Zbiory dla Andersona-Darlinga")
for kolumn in train_to_analise.drop(columns=categorical_columns):
    andersonTest = stats.anderson(train_to_analise[kolumn])
    for critical in andersonTest.critical_values:
        if andersonTest.statistic < critical:
            Zbiór_wartości_o_rozkładzie_normalnym.append([critical, kolumn])
            break
        else:
            Odrzucone_zbiory.append([critical, kolumn])

print()
print("Zbiory o rozkładzie normalnym")
print(Zbiór_wartości_o_rozkładzie_normalnym)
print()
print("Zbiory odrzucone o nie o rozkładzie normalnym")
print(Odrzucone_zbiory)
```

## Conclusion!!!

# Correlograms

```python
# Extracting numeric data from my main data
train_to_pearsonCorelogram = train_to_analise.drop(columns=categorical_columns)

# Calculating the correlation matrix
# The .corr() method calculates Pearson's correlation coefficient between each pair of columns in table d.
# The result is a new table (DataFrame) with dimensions, for example, 26x26, where each cell (i, j) contains 
# information about the correlation between column i and column j.
corr = train_to_pearsonCorelogram.corr()
```

e.g. ones_like(corr, dtype=bool): Creates a matrix with the same shape as corr, but filled with True values.
e.g.triu(...): The triu (from triangle upper) function takes this True matrix and returns its upper triangle. The values in the upper triangle (including the diagonal) remain True, and the rest become False.

As a result, the mask is a matrix of True/False values that will be used to “cover” the upper, unnecessary part of the heatmap.

```python
from matplotlib.pyplot import figure

mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(20, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
# Jestem świadomy ustawiania przedziału od 3 do -3 ponieważ dzięki temu moja pula kolorów będzie lepsza do analizy a gdyby była by pula kolorów od
# 1 do -1 to kolory wydawały by się bardzo podobne i nie było by widać dokładniejszych różnic.
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": .5})
```

Conclusion drawn from the correlogram analysis:

# Obliczanie Variance inflaction factor (VIF) dla kolumn numerycznych
Identyfikacja zmiennych numerycznych obciążonych problemem współliniowości (multicollinearity), czyli sytuacji, w której jedna zmienna jest silnie skorelowana z kombinacją innych zmiennych.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Remove categorical values because this function cannot handle them very well.
data_vif = train_to_analise.drop(columns=categorical_columns)

# Quick function mapping. I send the given value X.values instead of the usual x because I only need to send a matrix to the function, not a dataframe from pandas.
# So I send the variance_inflation_factor function one parameter of the matrix So I send variance_inflation_factor to the function, one parameter of the matrix
# and the second index of the calculated column. Thanks to the mapping, I can quickly obtain a table with all calculated 
# VIF values for each column
# We create a DataFrame with two columns right away.
data_vif = pd.DataFrame({
    "feature": data_vif.columns, # Kolumna z nazwami
    "VIF": [variance_inflation_factor(data_vif.values, i) for i in range(len(data_vif.columns))]
})

display(data_vif)
```

# Connecting using PCU
Objective: Reduction of dimensionality and elimination of multicollinearity between strongly correlated numerical features.

I identified groups of highly correlated features (e.g., garage area vs. number of parking spaces, ground floor area vs. basement area). Instead of removing one of them and losing potentially valuable information, I used PCA (Principal Component Analysis).

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

column_PCU = train_to_analise[['TotalBsmtSF','BsmtUnfSF','BsmtFinSF2','BsmtFinSF1']]
column_PCU_test = test_to_analise[['TotalBsmtSF','BsmtUnfSF','BsmtFinSF2','BsmtFinSF1']]

scaler = StandardScaler()
column_PCU = scaler.fit_transform(column_PCU)
column_PCU_test = scaler.fit_transform(column_PCU_test)

pca = PCA(n_components=1)
column_PCU = pca.fit_transform(column_PCU)
column_PCU_test = pca.fit_transform(column_PCU_test)

train_to_analise = train_to_analise.drop(columns=['TotalBsmtSF','BsmtUnfSF','BsmtFinSF2','BsmtFinSF1'])
train_to_analise['PC_added'] = column_PCU

test_to_analise = test_to_analise.drop(columns=['TotalBsmtSF','BsmtUnfSF','BsmtFinSF2','BsmtFinSF1'])
test_to_analise['PC_added'] = column_PCU_test

print(column_PCU)
```
