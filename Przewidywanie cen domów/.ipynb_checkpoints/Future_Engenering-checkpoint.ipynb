```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("./Zbiory danych do przetwarzania/train.csv")
test = pd.read_csv("./Zbiory danych do przetwarzania/test.csv")
```

```python
# podstawowe informacje o pobranych tabelach
print(f"Training set shape: {train.shape}")
print(f"Test set shape: {test.shape}")
print(f"Features: {train.shape[1]-1}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5))

# Original distribution
axes[0].hist(train['SalePrice'], bins=50, color='skyblue', edgecolor='black')
axes[0].set_title('SalePrice Distribution')
axes[0].set_xlabel('Sale Price')
axes[0].set_ylabel('Frequency')

# Log-transformed distribution
axes[1].hist(np.log1p(train['SalePrice']), bins=50, color='lightcoral', edgecolor='black')
axes[1].set_title('Log(SalePrice + 1) Distribution')
axes[1].set_xlabel('Log(Sale Price + 1)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

print("Log transformation makes the target variable more normally distributed")
```
